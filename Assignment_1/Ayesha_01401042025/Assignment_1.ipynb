```python
# Find Area of a Rectangle

length = float(input("Enter length: "))
width = float(input("Enter width: "))

area = length * width

print("Area of Rectangle =", area)
```

Input
Enter length: 10
Enter width: 5
Output
Area of Rectangle = 50.0

```python
# Find Simple Interest

principal = float(input("Enter Principal Amount: "))
rate = float(input("Enter Rate of Interest: "))
time = float(input("Enter Time (in years): "))

simple_interest = (principal * rate * time) / 100

print("Simple Interest =", simple_interest)
```

Input
Enter Principal Amount: 10000
Enter Rate of Interest: 5
Enter Time (in years): 2
Output
Simple Interest = 1000.0

```python
# Celsius to Fahrenheit Conversion

celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print("Temperature in Fahrenheit =", fahrenheit)
```

Input
Enter temperature in Celsius: 25
Output
Temperature in Fahrenheit = 77.0

```python
# Calculate Average of 3 Numbers

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

average = (num1 + num2 + num3) / 3

print("Average =", average)
```

Input
Enter first number: 10
Enter second number: 20
Enter third number: 30
Output
Average = 20.0

```python
# Find Square and Cube of a Number

number = float(input("Enter a number: "))

square = number ** 2
cube = number ** 3

print("Square =", square)
print("Cube =", cube)
```

Input
Enter a number: 4
Output
Square = 16.0
Cube = 64.0

```python
# Swap Two Numbers Without Third Variable

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Before Swapping:")
print("a =", a)
print("b =", b)

a, b = b, a

print("After Swapping:")
print("a =", a)
print("b =", b)
```

Input
Enter first number: 15
Enter second number: 25
Output
Before Swapping:
a = 15
b = 25

After Swapping:
a = 25
b = 15

```python
# Student Report Program
# This program takes student details,
# calculates total marks and percentage.

# Input student details
student_name = input("Enter Student Name: ")
roll_number = input("Enter Roll Number: ")

# Input marks
subject1 = float(input("Enter marks in Subject 1: "))
subject2 = float(input("Enter marks in Subject 2: "))
subject3 = float(input("Enter marks in Subject 3: "))
subject4 = float(input("Enter marks in Subject 4: "))
subject5 = float(input("Enter marks in Subject 5: "))

# Calculate total marks
total_marks = subject1 + subject2 + subject3 + subject4 + subject5

# Calculate percentage
percentage = total_marks / 5

# Display report
print("\nSTUDENT REPORT")
print("Student Name :", student_name)
print("Roll Number  :", roll_number)
print("Total Marks  :", total_marks)
print("Percentage   :", percentage, "%")
```

Input
Enter Student Name: Rahul Sharma
Enter Roll Number: 101
Enter marks in Subject 1: 85
Enter marks in Subject 2: 90
Enter marks in Subject 3: 78
Enter marks in Subject 4: 88
Enter marks in Subject 5: 92
Output
STUDENT REPORT
Student Name : Rahul Sharma
Roll Number  : 101
Total Marks  : 433.0
Percentage   : 86.6 %
